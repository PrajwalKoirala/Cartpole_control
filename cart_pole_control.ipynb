```python
from cartpole import CartPoleEnv
```

```python
import gym 
import numpy as np
import matplotlib.pyplot as plt
from time import time
from gym import wrappers
```

```python
env = CartPoleEnv()
```

```python

k = np.array([[ -3.1623,  -12.3185, -187.7909  ,-57.407]]) 
```

```python
new_state = env.reset()
count = 0;
done =   False                 
r = np.array([[2,0,0,0]]).T
angle = []
position = []
actions = []
while not done:
    angle.append(new_state[2])
    position.append(new_state[0])
    error = r - new_state.reshape((4,1))
    action = np.dot(k, error)
    actions.append(action[0])
    new_state, reward, done, info = env.step(action)      
    env.render()
    count+=1
    if count>600:
        break
env.close()                                             
```

```python
plt.plot(actions, label = 'force')
plt.xlabel('episodes')
plt.ylabel('force')
plt.legend()
plt.savefig("force.png")
```

```python
plt.plot(position, label = 'cart position')
plt.plot(np.zeros((len(position),1))+2,'r--', label = 'reference position')

plt.xlabel('episodes')
plt.ylabel('cart position')
plt.legend()
plt.savefig("position.png")
```

```python
plt.plot(angle, label = 'pole angle')
plt.plot(np.zeros((len(angle),1)),'r--', label = '0 radian')

plt.xlabel('episodes')
plt.ylabel('pole angle')
plt.legend()
plt.savefig("pole.png")
```
